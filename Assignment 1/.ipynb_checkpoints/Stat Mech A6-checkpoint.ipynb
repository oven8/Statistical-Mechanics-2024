The probability of getting a letter from the bigrams is,
\begin{equation}
P_B(a) = \sum_x P(ax)+\sum_y P(ya)
\end{equation}
The probability of a bigram given a letter is,
\begin{equation}
P(ab|a) = \frac{P(ab)}{P_B(a)}
\end{equation}
Then the probability of a letter occuring given a bigram is,
\begin{equation}
P(a|ab) = \frac{P(ab|a)P(a)}{P(ab)}
\end{equation}
Then the new probability of the letter occuring is,
\begin{equation}
P_N(a) = \sum_{xy} P(a|xy)P(xy)
\end{equation}
Note that the following possibilities can arise:
\begin{align}
P_N(a) &= \sum_{y} P(a|ay)P(ay)+\sum_{x} P(a|xa)P(xa)+P(a)\sum_{x,y\neq a} P(xy)\\
&= \sum_{y} P(a|ay)P(ay)+\sum_{x} P(a|xa)P(xa)+P(a)[1-P_B(a)]
\end{align}

$P_B(a) = \sum_x P(ax)+\sum_y P(ya)$

```python
Sum_pr_bigram_turk = [0]*len(alp)
for i in range(len(alp)):
    for j in range(np.shape(Prob_bigram_turk)[0]):
        if (Prob_bigram_turk.iloc[j,0][0] == alp[i]) or (Prob_bigram_turk.iloc[j,0][1] == alp[i]):
            Sum_pr_bigram_turk[i] += Pr_turk_bigram_array[j]    
            
Sum_pr_bigram_turk = np.array(Sum_pr_bigram_turk)
Sum_pr_bigram_turk
```

$P(ab|a) = \frac{P(ab)}{P_B(a)}$

```python
Pr_ab_given_a_turk = [0]*np.shape(Prob_bigram_turk)[0]
for i in range(np.shape(Prob_bigram_turk)[0]):
    for j in range(len(alp)):
        if Prob_bigram_turk.iloc[i,0][0] == alp[j]:
            k = alp.index(Prob_bigram_turk.iloc[i,0][0])
            Pr_ab_given_a_turk[i] = Pr_turk_bigram_array[i]/Sum_pr_bigram_turk[j]
            
Pr_ab_given_a_turk = np.array(Pr_ab_given_a_turk)
Pr_ab_given_a_turk
```

$P(a|ab) = \frac{P(ab|a)P(a)}{P(ab)}$

```python
Pr_a_given_ab_turk = [0]*len(alp)
for i in range(np.shape(Prob_bigram_turk)[0]):
    for j in range(len(alp)):
        if Prob_bigram_turk.iloc[i,0][0] == alp[j]:
            k = alp.index(Prob_bigram_turk.iloc[i,0][0])
            Pr_a_given_ab_turk[j] = Pr_ab_given_a_turk[i]*Pr_turk_array[j]/Pr_turk_bigram_array[i]
            
Pr_a_given_ab_turk = np.array(Pr_ab_given_a_turk)
Pr_a_given_ab_turk
```

```python
Pr_ba_given_a_turk = [0]*np.shape(Prob_bigram_turk)[0]
for i in range(np.shape(Prob_bigram_turk)[0]):
    for j in range(len(alp)):
        if Prob_bigram_turk.iloc[i,0][0] == alp[j]:
            k = alp.index(Prob_bigram_turk.iloc[i,0][1])
            Pr_ba_given_a_turk[i] = Pr_turk_bigram_array[i]/Sum_pr_bigram_turk[k]
            
Pr_ba_given_a_turk = np.array(Pr_ba_given_a_turk)
Pr_ba_given_a_turk
```
